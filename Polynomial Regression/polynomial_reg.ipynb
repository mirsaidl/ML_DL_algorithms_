```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
arr = np.arange(10,2000,15)
arr_root = np.sqrt(arr) 
arr_sq = [x**2 for x in arr]

fig, ax = plt.subplots(2, 1, figsize=(8, 6))

sns.scatterplot(ax=ax[0], x=arr, y=arr_root, color='r')
sns.scatterplot(ax=ax[1], x=arr, y=arr_sq)

plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generate some random data
np.random.seed(0)
x = 2 * np.pi * np.random.rand(100, 1)
y = np.sin(x) + 0.1 * np.random.randn(100, 1)

# Plot the data
plt.scatter(x, y, s=10, label='Data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression Example')
plt.legend()
plt.show()
```

```python
# Create polynomial features
degree = 12  # Degree of the polynomial
poly_features = PolynomialFeatures(degree=degree, include_bias=False)
x_poly = poly_features.fit_transform(x)

# Fit a linear regression model
model = LinearRegression()
model.fit(x_poly, y)

# Predictions
x_new = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
x_new_poly = poly_features.transform(x_new)
y_pred = model.predict(x_new_poly)

# Plot the data and the fitted curve
plt.scatter(x, y, s=10, label='Data')
plt.plot(x_new, y_pred, color='red', label='Polynomial Regression (Degree {})'.format(degree))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression Example')
plt.legend()
plt.show()
```
